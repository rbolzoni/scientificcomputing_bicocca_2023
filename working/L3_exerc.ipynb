## Q1: Planetary positions

The distances of the planets from the Sun (technically, their semi-major axes) are:

```python
import numpy as np
a = np.array([0.39, 0.72, 1.00, 1.52, 5.20, 9.54, 19.22, 30.06, 39.48])
P = np.array([0.24, 0.62, 1.00, 1.88, 11.86, 29.46, 84.01, 164.8, 248.09])
names = ["Mercury", "Venus", "Earth", "Mars", "Jupiter", "Saturn", 
         "Uranus", "Neptune", "Pluto"]
```

The period vs semi-axis is plotted in a log-log scale. Zip() allows to assign every name to its corresponding coordinate (the period is multiplied by 1.1 so that the name is slightly shifted to the right and doesn't cover the markers)

```python
import matplotlib.pyplot as plt

plt.scatter(a,P,marker='*')
plt.xscale('log')
plt.yscale('log')
c=sorted(zip(a,P,names))
for l in range(a.size):
    plt.text(c[l][0]*1.1,c[l][1],c[l][2]) 
```

## Q4: Subplots

Create 3 axes vertically, and do the following:

* Define a new numpy array `f` initialized to a function of your choice.
* Plot f in the top axes
* Compute a numerical derivative of `f`,
   $$ f' = \frac{f_{i+1} - f_i}{\Delta x}$$
  and plot this in the middle axes
* Do this again, this time on $f'$ to compute the second derivative and plot that in the bottom axes


```python
import numpy as np
import scipy.special as scp
import matplotlib.pyplot as plt

n=50
x=np.linspace(0,2*np.pi,n)
fun = np.sin(x)


fig, ax = plt.subplots(3,1, sharex='col')
ax[0].plot(x,fun)
ax[0].set_ylabel('sen(x)')


der = np.zeros(n-1)
for l in range(0,n-1):
    der[l]=(fun[l+1]-fun[l])/(x[l+1]-x[l])
ax[1].plot(x[0:n-1],der)
ax[1].set_ylabel('Derivative')


der2 = np.zeros(n-2)
for l in range(0,n-2):
    der2[l]=(der[l+1]-der[l])/(x[l+1]-x[l])
ax[2].plot(x[0:n-2],der2)
ax[2].set_ylabel('Second Derivative')
ax[2].set_xlabel('x')
```
